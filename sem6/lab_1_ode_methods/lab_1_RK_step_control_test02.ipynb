<h2> Демонстрация алгоритма контроля шага (правило Рунге) </h2>

Для системы: 

$f_{actual}(t) = e^{\sin^2 t}$

$\dfrac{d f_{actual}(t)}{d t} = f_{sys}(t)$

$ 
    \begin{dcases}
    f_{sys}(t) = 2 e^{\sin^2 t} \cos t \sin t, \\
    f_{sys}(0) = 1.
    \end{dcases} 
$

```python
import numpy as np
import pandas as pd

import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
from plotly.subplots import make_subplots

init_notebook_mode(connected=True)

fout_dir = "./output/test02_for_runge_rule/"
file_name_2 = "test02_tau_1e-2"


```

```python

def draw_plots_plotly(t, u, y, tau):
    
    layout = go.Layout(autosize=False, width=800, height=600)

    fig = go.Figure(layout=layout)

    fig = make_subplots(
    rows=1, cols=2,    
    subplot_titles=("Plot of Time against Y", "Plot of Time against TAU")
    )
    

    fig.add_trace(
        go.Scatter(x=t, y=u, mode="lines", name="Plot of Time against F_actual"), 
        row=1, col=1,

    )

    fig.add_trace(
        go.Scatter(x=t, y=y, mode="lines", name="Plot of Time against F_found"), 
        row=1, col=1,
    )

    fig.add_trace(
        go.Scatter(x=t, y=tau, mode="lines", name="Plot of Time against Tau"),  
        row=1, col=2,
    )
    
    fig.update_layout(
        title="""Plot of X against Y""",
        showlegend=True,
    )
    fig.update_xaxes(
        showline=True,
        linewidth=1,
        linecolor="black",
        minor=dict(ticklen=5, tickcolor="black", showgrid=True),
    )
    fig.update_yaxes(
        showline=True,
        linewidth=1,
        linecolor="black",
        minor=dict(ticklen=5, tickcolor="black", showgrid=True),
    )
    
    iplot(fig)

```

<H2> Метод Рунге-Кутты 2 порядка (с меняющимся шагом)

```python
method_name = "_ode_RK2_vary_step_output"

df = pd.read_csv(fout_dir + file_name_2 + method_name + ".csv")


def u_func(t):
    return np.exp(np.sin(t) * np.sin(t))

t = df["time"]
y = df["u0"]
tau = df["tau"]
u = u_func(t)

draw_plots_plotly(t, y, u, tau)
```

<H2> Метод Рунге-Кутты 4 порядка (с меняющимся шагом)

```python
method_name = "_ode_RK4_vary_step_output"

df = pd.read_csv(fout_dir + file_name_2 + method_name + ".csv")


def u_func(t):
    return np.exp(np.sin(t) * np.sin(t))

t = df["time"]
y = df["u0"]
tau = df["tau"]
u = u_func(t)

draw_plots_plotly(t, y, u, tau)
```
