<H1>Лабораторная работа №1 (ode)

```python
import numpy as np
import pandas as pd
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True)

fout_dir = "./output/test0_for_approximation_order/"
file_name_2 = "test0_tau_1e-2"
file_name_3 = "test0_tau_1e-3"
file_name_4 = "test0_tau_1e-4"


```

```python

def draw_plots_plotly(t, u, y):
    
    layout = go.Layout(autosize=False, width=800, height=600)

    fig = go.Figure(layout=layout)

    fig.add_traces(
        [
            go.Scatter(x=t, y=u, mode="lines", name="Plot of Time against X"), 
            go.Scatter(x=t, y=y, mode="lines", name="Plot of Time against Y"), 
        ]

    )
    
    fig.update_layout(
        title="""Plot of X against Y""",
        #showlegend=True,
    )
    fig.update_xaxes(
        showline=True,
        linewidth=1,
        linecolor="black",
        minor=dict(ticklen=5, tickcolor="black", showgrid=True),
    )
    fig.update_yaxes(
        showline=True,
        linewidth=1,
        linecolor="black",
        minor=dict(ticklen=5, tickcolor="black", showgrid=True),
    )
    
    iplot(fig)

```

<H2> Явный метод Эйлера

```python
method_name = "_ode_explicit_Euler_output"

df = pd.read_csv(fout_dir + file_name_2 + method_name + ".csv")


def u_func(t):
    return np.exp(np.sin(t)) - np.cos(t) + 0.5


t = df["time"]
y = df["u0"]
u = u_func(t)


draw_plots_plotly(t, y, u)

```
