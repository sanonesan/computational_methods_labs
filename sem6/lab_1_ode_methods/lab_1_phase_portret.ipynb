<h1> Фазовые портреты систем

```python
import numpy as np
import plotly.figure_factory as ff
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

init_notebook_mode(connected=True)
```

<h2> Маятник

```python
k = 20
m = 0.3
w = k/m


x = np.linspace(-3, 3, 1000)
y = np.linspace(-3, 3, 1000)
X, Y = np.meshgrid(x, y)
u = Y
v = -w * X



layout = go.Layout(autosize=False, width=1200, height=400)

fig = ff.create_streamline(x, y, u, v)
fig.layout = layout

fig.update_layout(
    title="""Phase portrait""",
    #showlegend=True,
)
fig.update_xaxes(
    showline=True,
    linewidth=1,
    linecolor="black",
    minor=dict(ticklen=5, tickcolor="black", showgrid=True),
)
fig.update_yaxes(
    showline=True,
    linewidth=1,
    linecolor="black",
    minor=dict(ticklen=5, tickcolor="black", showgrid=True),
)

fig.write_image("./media/" + "Маятник" + "_streamplot" + ".pdf")

iplot(fig)
```

<h2> Вариант 4 (диссипативная система)

```python
a = 0.1
mu = 0.1
w = 0.25
nu = 0.15

t = 0


def func_H(x, y): 
    return (x**2 + y**2)**2 - 2 * a**2 * (x**2 + y**2)

def func_exact_deriv_dH_dx_i(x, y, i):
    if(i == 0):
        return -4 * a**2 * x + 4 * x * (x**2 + y**2)
    if(i == 1):
        return 4 * a**2 * y + 4 * y * (x**2 + y**2)
    

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x, y)

u = func_exact_deriv_dH_dx_i(X, Y, 1) - mu * func_H(X, Y) * func_exact_deriv_dH_dx_i(X, Y, 0) + nu * Y * np.sin(w * t)
v = -func_exact_deriv_dH_dx_i(X, Y, 0) - mu * func_H(X, Y) * func_exact_deriv_dH_dx_i(X, Y, 1) + nu * Y * np.sin(w * t);


layout = go.Layout(autosize=False, width=800, height=800)

fig = ff.create_streamline(x, y, u, v)
fig.layout = layout

fig.update_layout(
    title="""Phase portrait""",
    #showlegend=True,
)
fig.update_xaxes(
    showline=True,
    linewidth=1,
    linecolor="black",
    minor=dict(ticklen=5, tickcolor="black", showgrid=True),
)
fig.update_yaxes(
    showline=True,
    linewidth=1,
    linecolor="black",
    minor=dict(ticklen=5, tickcolor="black", showgrid=True),
)

fig.write_image("./media/" + "Маятник" + "_streamplot" + ".pdf")

iplot(fig)
```
