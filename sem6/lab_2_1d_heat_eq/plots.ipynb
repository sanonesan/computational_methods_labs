```python
from matplotlib import pyplot as plt
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
import numpy as np
init_notebook_mode(connected=True)
out_dir = "./output"
file_name = "/"
```

```python
import pandas as pd

def draw_plot(out_dir):

    layout = go.Layout(autosize=False, width=800, height=600)

    fig = go.Figure(layout=layout)

    method_name = "_1d_heat_eq_output"

    # traces = []
    
    df = pd.read_csv(out_dir + file_name + method_name + ".csv")
    
    t = df["time"]
    x = list(df.iloc[[0]].values.flatten().tolist())
    x.pop(0)    
    z = []
    for i in x:
        z.append(t[0])
    traces = []
    for i in range(1,5):
        y = list(df.iloc[[i]].values.flatten().tolist())
        y.pop(0)  
        
        for j in range(len(y)):
            z[j] = t[i]  

        traces.append(go.Scatter3d(x=x, y=z, z=y, mode="lines", name="Plot"))
            
    fig.add_traces(traces)
    #fig = go.Figure(data=go.Scatter3d(x=x, y=z,z=y, mode='lines'))
    # fig.update_layout(
    #     title="""profile""",
    #     showlegend=False,
    # )
    # fig.update_xaxes(
    #     showline=True,
    #     linewidth=1,
    #     linecolor="black",
    #     minor=dict(ticklen=5, tickcolor="black", showgrid=True),
    # )
    # fig.update_yaxes(
    #     showline=True,
    #     linewidth=1,
    #     linecolor="black",
    #     minor=dict(ticklen=5, tickcolor="black", showgrid=True),
    # )

    fig.show()

```

```python
out_dir = "./output"
draw_plot(out_dir)
```
